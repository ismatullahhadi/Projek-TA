```python
# Import Library
import numpy as np # linear algebra
import pandas as pd # data processing, pembacaan file excel
import matplotlib.pyplot as plt # untuk plotting
```

```python
# Pembacaan File Data Polusi Udara ISPU Excel
DKI1 = pd.read_excel("./Data/Data ISPU.xlsx", sheet_name="DKI1")
DKI2 = pd.read_excel("./Data/Data ISPU.xlsx", sheet_name="DKI2")
DKI3 = pd.read_excel("./Data/Data ISPU.xlsx", sheet_name="DKI3")
DKI4 = pd.read_excel("./Data/Data ISPU.xlsx", sheet_name="DKI4")
DKI5 = pd.read_excel("./Data/Data ISPU.xlsx", sheet_name="DKI5")
```

```python
# Menampilkan 5 Data pertama DKI1
DKI1.head()
```

```python
# Hanya menampilkan data zat yang diukur
DKI1_Data = DKI1.loc[:, ~DKI1.columns.isin(['Tanggal', 'Wilayah', 'Critical', 'Kategori'])]
DKI2_Data = DKI2.loc[:, ~DKI2.columns.isin(['Tanggal', 'Wilayah', 'Critical', 'Kategori'])]
DKI3_Data = DKI3.loc[:, ~DKI3.columns.isin(['Tanggal', 'Wilayah', 'Critical', 'Kategori'])]
DKI4_Data = DKI4.loc[:, ~DKI4.columns.isin(['Tanggal', 'Wilayah', 'Critical', 'Kategori'])]
DKI5_Data = DKI5.loc[:, ~DKI5.columns.isin(['Tanggal', 'Wilayah', 'Critical', 'Kategori'])]
DKI1_Data
```

```python
# KNN Imputer
from sklearn.impute import KNNImputer # import KNN Imputer untuk melakukan pengisian data yang hilang
```

```python
# Impute data ke dalam array
impute_knn = KNNImputer(n_neighbors=5)
DKI1_Array = impute_knn.fit_transform(DKI1_Data)
DKI2_Array = impute_knn.fit_transform(DKI2_Data)
DKI3_Array = impute_knn.fit_transform(DKI3_Data)
DKI4_Array = impute_knn.fit_transform(DKI4_Data)
DKI5_Array = impute_knn.fit_transform(DKI5_Data)
```

```python
# Import ke CSV
DKI1_Impute = pd.DataFrame(DKI1_Array)
DKI2_Impute = pd.DataFrame(DKI2_Array)
DKI3_Impute = pd.DataFrame(DKI3_Array)
DKI4_Impute = pd.DataFrame(DKI4_Array)
DKI5_Impute = pd.DataFrame(DKI5_Array)

DKI1_Impute.to_csv('.\Data\Impute\DKI1_Impute.csv')
DKI2_Impute.to_csv('.\Data\Impute\DKI2_Impute.csv')
DKI3_Impute.to_csv('.\Data\Impute\DKI3_Impute.csv')
DKI4_Impute.to_csv('.\Data\Impute\DKI4_Impute.csv')
DKI5_Impute.to_csv('.\Data\Impute\DKI5_Impute.csv')
```
