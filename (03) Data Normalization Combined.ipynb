```python
# Pembacaan File Data Polusi Udara ISPU Excel
DKI1 = pd.read_excel("./Data/Data ISPU - Combined.xlsx", sheet_name="DKI1")
DKI2 = pd.read_excel("./Data/Data ISPU - Combined.xlsx", sheet_name="DKI2")
DKI3 = pd.read_excel("./Data/Data ISPU - Combined.xlsx", sheet_name="DKI3")
DKI4 = pd.read_excel("./Data/Data ISPU - Combined.xlsx", sheet_name="DKI4")
DKI5 = pd.read_excel("./Data/Data ISPU - Combined.xlsx", sheet_name="DKI5")
```
