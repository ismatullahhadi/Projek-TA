```python
# Import Library
import numpy as np # linear algebra
import pandas as pd # data processing, pembacaan file excel
import matplotlib.pyplot as plt # untuk plotting
from sklearn import preprocessing # preprocessing untuk normalisasi
```

```python
# Pembacaan File Data Polusi Udara ISPU Excel
DKI1 = pd.read_excel("./Data/Impute/Data ISPU - Impute.xlsx", sheet_name="DKI1")
DKI2 = pd.read_excel("./Data/Impute/Data ISPU - Impute.xlsx", sheet_name="DKI2")
DKI3 = pd.read_excel("./Data/Impute/Data ISPU - Impute.xlsx", sheet_name="DKI3")
DKI4 = pd.read_excel("./Data/Impute/Data ISPU - Impute.xlsx", sheet_name="DKI4")
DKI5 = pd.read_excel("./Data/Impute/Data ISPU - Impute.xlsx", sheet_name="DKI5")
```

```python
# Menampilkan 5 Data pertama DKI1
DKI1.head()
```

```python
# Hanya menampilkan data zat yang diukur
DKI1_Data = DKI1.loc[:, ~DKI1.columns.isin(['Tanggal', 'Wilayah', 'Critical', 'Kategori'])]
DKI2_Data = DKI2.loc[:, ~DKI2.columns.isin(['Tanggal', 'Wilayah', 'Critical', 'Kategori'])]
DKI3_Data = DKI3.loc[:, ~DKI3.columns.isin(['Tanggal', 'Wilayah', 'Critical', 'Kategori'])]
DKI4_Data = DKI4.loc[:, ~DKI4.columns.isin(['Tanggal', 'Wilayah', 'Critical', 'Kategori'])]
DKI5_Data = DKI5.loc[:, ~DKI5.columns.isin(['Tanggal', 'Wilayah', 'Critical', 'Kategori'])]
DKI1_Data.head()
```

```python
print('MAX      : ', max(DKI1['NO2']))
print('MIN      : ', min(DKI1['NO2']))
```

```python
# Normalisasi PM10
x_array = np.array(DKI1_Data['PM10'])
normalized_arr = preprocessing.normalize([x_array])
print(normalized_arr)
```

```python
# Normalisasi Data DKI1 sampai DKI5
scaler = preprocessing.MinMaxScaler()
names1 = DKI1_Data.columns
names2 = DKI2_Data.columns
names3 = DKI3_Data.columns
names4 = DKI4_Data.columns
names5 = DKI5_Data.columns
d1 = scaler.fit_transform(DKI1_Data)
d2 = scaler.fit_transform(DKI2_Data)
d3 = scaler.fit_transform(DKI3_Data)
d4 = scaler.fit_transform(DKI4_Data)
d5 = scaler.fit_transform(DKI5_Data)
DKI1_Normalization = pd.DataFrame(d1, columns=names1)
DKI2_Normalization = pd.DataFrame(d2, columns=names2)
DKI3_Normalization = pd.DataFrame(d3, columns=names3)
DKI4_Normalization = pd.DataFrame(d4, columns=names4)
DKI5_Normalization = pd.DataFrame(d5, columns=names5)
DKI1_Normalization.head()
```

```python
# Penyimpanan data yang sudah normalisasi ke dalam file csv
DKI1_Normalization.to_csv('./Data/Normalization/DKI1_Normalization.csv')
DKI2_Normalization.to_csv('./Data/Normalization/DKI2_Normalization.csv')
DKI3_Normalization.to_csv('./Data/Normalization/DKI3_Normalization.csv')
DKI4_Normalization.to_csv('./Data/Normalization/DKI4_Normalization.csv')
DKI5_Normalization.to_csv('./Data/Normalization/DKI5_Normalization.csv')
```
